```python
import pandas as pd
df_2023 = pd.read_stata("nsch_2023e_topical.dta")
df_2023 = df_2023.iloc[:, 5:]

df_2016 = pd.read_stata("nsch_2016e_topical.dta")

df_2016 = df_2016.iloc[:, 5:]

#df_2016.columns = df_2023.columns 

df_combined = pd.concat([df_2023, df_2016], ignore_index=True)

df_combined.to_csv("combo_data.csv", index=False)

print("Conversion complete: .dta goes into .csv")
```

```python
import pandas as pd
df = pd.read_csv("combo_data.csv")
 
df.columns.tolist()
```

```python
num_rows, num_columns = df.shape

# Print the number of rows
print(f"The dataset contains {num_rows} rows and {num_columns} columns.")

variables_to_clean = [
    "screentime", "k2q31a", "k7q04r_r", "grades", # Correlational Analysis
    "sc_age_years", "sc_sex", "physactiv", "sc_race_r"  # Logistic Regression
]

df_cleaned = df.dropna(subset=variables_to_clean)

df_cleaned.to_csv("ADHD_data.csv", index=False)
num_rows, num_columns = df_cleaned.shape
print(f"The dataset now contains {num_rows} rows and {num_columns} columns.")
```

```python
null_counts = df[variables_to_clean].isnull().sum()

# Print the number of null values in each variable
print("Number of null values in each variable:")
print(null_counts)
```
